```python
import pandas as pd
import numpy as np
from math import radians, sin, cos, sqrt, atan2
from pathlib import Path

# =====================================================
# CONTAINER CANDIDATES (rule-based)
# Input : all_merged_data_*.csv
# Output: container_candidates.csv
# =====================================================

IN_FILE  = Path("all_merged_data.csv")   # bunu all_merged_data.csv yapabilirsin
OUT_FILE = Path("container_candidates.csv")

# -------------------------
# PARAMETRELER
# -------------------------
STOP_LABELS = {"Duran", "Rölanti Alarmı", "Duraklama Alarmı"}

TIME_GAP_MIN = 20
DIST_GAP_M   = 80

MIN_DUR_MIN_DEFAULT = 10

Q_LOW, Q_HIGH = 0.01, 0.99

BONUS_IDLE_ALARM = 2.0
BONUS_STOP_ALARM = 1.5
REPEAT_BONUS_2DAYS = 1.0
REPEAT_BONUS_4DAYS = 2.0

# -------------------------
# Yardımcılar
# -------------------------
def haversine_m(lat1, lon1, lat2, lon2):
    R = 6371000.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    return 2 * R * atan2(sqrt(a), sqrt(1 - a))

def pick_col(df, candidates):
    lower_map = {c.lower(): c for c in df.columns}
    for cand in candidates:
        if cand.lower() in lower_map:
            return lower_map[cand.lower()]
    return None

def robust_datetime(df):
    ts = pick_col(df, ["timestamp", "ts", "datetime", "date_time", "tarih_saat"])
    if ts:
        return pd.to_datetime(df[ts], errors="coerce", dayfirst=True)

    dcol = pick_col(df, ["date", "tarih"])
    tcol = pick_col(df, ["time", "saat"])
    if dcol and tcol:
        return pd.to_datetime(
            df[dcol].astype(str) + " " + df[tcol].astype(str),
            errors="coerce",
            dayfirst=True
        )

    if dcol:
        return pd.to_datetime(df[dcol], errors="coerce", dayfirst=True)

    return pd.Series([pd.NaT] * len(df))

def duration_col_to_minutes(series: pd.Series) -> pd.Series:
    """
    Süre kolonunu dakikaya çevirir.
    Destek:
      - HH:MM:SS / MM:SS string (pd.to_timedelta)
      - numerik (saniye veya dakika heuristiği)
    """
    s = series.copy()

    # string gibi görünenleri yakala
    s_str = s.astype(str)
    has_colon = s_str.str.contains(":", regex=False, na=False)

    out = pd.Series(np.nan, index=s.index, dtype="float64")

    # HH:MM:SS tarzı -> timedelta
    if has_colon.any():
        td = pd.to_timedelta(s_str.where(has_colon), errors="coerce")
        out.loc[has_colon] = td.dt.total_seconds() / 60.0

    # kalanları numerik parse et
    num = pd.to_numeric(s.where(~has_colon), errors="coerce")
    if num.notna().any():
        # büyük değerleri saniye varsay (örn 600 -> 10 dk)
        # küçük değerler dakika olabilir
        # bunu max üzerinden kullanacağız, dolayısıyla basit heuristik yeterli
        out.loc[num.notna()] = np.where(num > 120, num / 60.0, num)

    return out.fillna(0.0)

# -------------------------
# 1) Oku + kolonları bul
# -------------------------
df = pd.read_csv(IN_FILE, low_memory=False)

col_vehicle = pick_col(df, ["vehicle_id", "arac_id", "arac", "plaka", "device_id"])
col_lat     = pick_col(df, ["lat", "latitude", "enlem", "Enlem", "street_latitude"])
col_lon     = pick_col(df, ["lon", "lng", "longitude", "boylam", "Boylam", "street_longitude"])
col_desc    = pick_col(df, ["Açıklama", "aciklama", "description", "durum", "Durum"])

# duration kolonları (senin dosyada boşluklu)
col_dur_stop = pick_col(df, [
    "Duraklama Süresi", "duraklama suresi",
    "duraklama_suresi", "duraklama_süresi",
    "stop_duration", "duraklama_sure"
])
col_dur_idle = pick_col(df, [
    "Rölanti Süresi", "rolanti suresi",
    "rolanti_suresi", "rölanti_suresi",
    "idle_duration", "rolanti_sure"
])

if col_vehicle is None or col_lat is None or col_lon is None or col_desc is None:
    raise ValueError(
        "Gerekli kolonlar bulunamadı.\n"
        f"vehicle_id: {col_vehicle}\nlat: {col_lat}\nlon: {col_lon}\nAçıklama: {col_desc}\n"
        f"Mevcut kolonlar: {list(df.columns)}"
    )

df["_dt"] = robust_datetime(df)

df[col_lat] = pd.to_numeric(df[col_lat], errors="coerce")
df[col_lon] = pd.to_numeric(df[col_lon], errors="coerce")
df = df.dropna(subset=[col_lat, col_lon]).copy()

df["_desc"] = (
    df[col_desc].astype(str)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

# -------------------------
# 2) Koordinat outlier kırp
# -------------------------
lat_q1, lat_q2 = df[col_lat].quantile(Q_LOW), df[col_lat].quantile(Q_HIGH)
lon_q1, lon_q2 = df[col_lon].quantile(Q_LOW), df[col_lon].quantile(Q_HIGH)
df = df[(df[col_lat].between(lat_q1, lat_q2)) & (df[col_lon].between(lon_q1, lon_q2))].copy()

# -------------------------
# 3) Stop aday satırlar
# -------------------------
cand = df[df["_desc"].isin(STOP_LABELS)].copy()

cand["_dt_filled"] = cand["_dt"]
if cand["_dt_filled"].isna().all():
    cand["_dt_filled"] = pd.to_datetime("1970-01-01") + pd.to_timedelta(cand.index, unit="s")

cand = cand.sort_values([col_vehicle, "_dt_filled"]).reset_index(drop=True)

# -------------------------
# 4) Segmentleme
# -------------------------
segments = []
for vid, g in cand.groupby(col_vehicle, sort=False):
    g = g.sort_values("_dt_filled").reset_index(drop=True)
    if len(g) == 0:
        continue

    seg_start_idx = 0
    for i in range(1, len(g)):
        dt_prev = g.loc[i-1, "_dt_filled"]
        dt_cur  = g.loc[i, "_dt_filled"]
        time_gap_min = (dt_cur - dt_prev).total_seconds() / 60.0

        d_m = haversine_m(
            g.loc[i-1, col_lat], g.loc[i-1, col_lon],
            g.loc[i, col_lat],   g.loc[i, col_lon]
        )

        if (time_gap_min > TIME_GAP_MIN) or (d_m > DIST_GAP_M):
            segments.append(g.iloc[seg_start_idx:i].copy())
            seg_start_idx = i

    segments.append(g.iloc[seg_start_idx:len(g)].copy())

# -------------------------
# 5) Segment özetleri -> container adayları
# -------------------------
rows = []
for seg in segments:
    if seg.empty:
        continue

    vid = seg[col_vehicle].iloc[0]
    start_time = seg["_dt_filled"].iloc[0]
    end_time   = seg["_dt_filled"].iloc[-1]

    duration_min_dt = (end_time - start_time).total_seconds() / 60.0
    if pd.isna(duration_min_dt) or duration_min_dt < 0:
        duration_min_dt = 0.0

    # süre kolonları varsa: SUM değil MAX (loglarda genelde birikmiş süre)
    dur_candidates = [duration_min_dt]

    for c in [col_dur_stop, col_dur_idle]:
        if c and c in seg.columns:
            mins = duration_col_to_minutes(seg[c])
            dur_candidates.append(float(mins.max()))

    duration_min = max(dur_candidates)

    lat = float(np.median(seg[col_lat].values))
    lon = float(np.median(seg[col_lon].values))

    aciklama = seg["_desc"].mode().iloc[0] if not seg["_desc"].mode().empty else str(seg["_desc"].iloc[0])

    has_idle_alarm = (seg["_desc"] == "Rölanti Alarmı").any()
    has_stop_alarm = (seg["_desc"] == "Duraklama Alarmı").any()

    min_dur = 0.0 if (has_idle_alarm or has_stop_alarm) else float(MIN_DUR_MIN_DEFAULT)
    if duration_min < min_dur:
        continue

    score = 0.0
    score += min(duration_min / 10.0, 6.0)
    if has_idle_alarm:
        score += BONUS_IDLE_ALARM
    if has_stop_alarm:
        score += BONUS_STOP_ALARM

    rows.append({
        "vehicle_id": vid,
        "date": pd.to_datetime(start_time).date().isoformat() if pd.notna(start_time) else None,
        "start_time": str(start_time),
        "end_time": str(end_time),
        "lat": round(lat, 7),
        "lon": round(lon, 7),
        "aciklama": aciklama,
        "duration_min": round(float(duration_min), 3),
        "score_base": round(float(score), 3),
        "n_rows": int(len(seg)),
    })

cand_seg = pd.DataFrame(rows)

if cand_seg.empty:
    cand_seg.to_csv(OUT_FILE, index=False)
    print("⚠ Hiç aday segment üretilemedi. Eşikler çok sert olabilir.")
    print(f"Çıktı yazıldı: {OUT_FILE.name}")
    raise SystemExit()

# -------------------------
# 6) Tekrarlama bonusu
# -------------------------
cand_seg["cell_id"] = (
    cand_seg["lat"].round(4).astype(str) + "_" + cand_seg["lon"].round(4).astype(str)
)

cell_days = (
    cand_seg.groupby("cell_id")["date"]
    .nunique()
    .rename("unique_days_at_cell")
    .reset_index()
)

cand_seg = cand_seg.merge(cell_days, on="cell_id", how="left")

def repeat_bonus(n_days):
    if n_days >= 4:
        return REPEAT_BONUS_4DAYS
    if n_days >= 2:
        return REPEAT_BONUS_2DAYS
    return 0.0

cand_seg["repeat_bonus"] = cand_seg["unique_days_at_cell"].apply(repeat_bonus)
cand_seg["score"] = (cand_seg["score_base"] + cand_seg["repeat_bonus"]).round(3)

# -------------------------
# 7) Çıktı
# -------------------------
out = cand_seg[[
    "lat", "lon", "aciklama", "duration_min", "score",
    "vehicle_id", "date", "start_time", "end_time",
    "unique_days_at_cell", "cell_id", "n_rows"
]].sort_values(["score", "duration_min"], ascending=[False, False])

out.to_csv(OUT_FILE, index=False)

print("✅ container_candidates.csv üretildi.")
print(f"-> {OUT_FILE.name}")
print(f"Aday sayısı: {len(out)}")
print("Top 10 aday:")
print(out.head(10))
```
